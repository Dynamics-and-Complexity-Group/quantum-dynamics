## No tunneling (Figure 3)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Read the two CSV files
file1 = "../data/Time_series_Energy_E_9.0_x_in_0.5.csv"
file2 = "../data/Time_series_Energy_E_13.678_x_in_5.5.csv"

# Load data from CSV files
data1 = pd.read_csv(file1)
data2 = pd.read_csv(file2)

# Extract time, x, and v from both files
time_win1 = data1['time'].values
time_win2 = data2['time'].values
x_vals1 = data1['x'].values
v_vals1 = data1['v'].values
x_vals2 = data2['x'].values
v_vals2 = data2['v'].values

# Enable LaTeX rendering
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rc('font', size=12)  # Increased font size for tick labels
plt.rc('xtick', labelsize=12); plt.rc('ytick', labelsize=12)  # Increase font size of tick labels (numbers) on x and y axes
# Create figure
fig = plt.figure(figsize=(8, 3))  # Wider figure for rectangular subplots
gs = gridspec.GridSpec(2, 2, height_ratios=[1, 1], width_ratios=[1, 1], hspace=0.0)

# Left panel: x and v values (stacked subplots with shared x-axis)
ax1_top = fig.add_subplot(gs[0, 0])
ax1_bottom = fig.add_subplot(gs[1, 0], sharex=ax1_top)

# Top left: x from file1 (E=7.4, x_in=0.1)
ax1_top.plot(time_win1, x_vals1, color='#00802b')
ax1_top.axhline(y=0.0, color="gray", linestyle="--")
ax1_top.axhline(y=3.604127443407402, color="red", alpha=0.5, linestyle="--")
ax1_top.axhline(y=8.160578438945539, color="gray", linestyle="--")
ax1_top.text(0.03, 0.95, '(a)', transform=ax1_top.transAxes, fontsize=12, va='top')
ax1_top.set_ylabel(r'$\langle x \rangle$', fontsize=12)
ax1_top.margins(x=0.05, y=0.05)  # Add padding to ensure text visibility
ax1_top.set_ylim([-1, 11.0])
# Bottom left: v from file1 (E=7.4, x_in=0.1)
ax1_bottom.plot(time_win1, v_vals1, color='#00802b')
ax1_bottom.text(0.03, 0.95, '(c)', transform=ax1_bottom.transAxes, fontsize=12, va='top')
ax1_bottom.set_ylabel(r'$\mathrm{Var}(x)$', fontsize=12)
ax1_bottom.set_xlabel(r'$t$', fontsize=12)
ax1_bottom.margins(x=0.05, y=0.05)  # Add padding to ensure text visibility
ax1_bottom.set_ylim([-0.5, 5.8])

# Right panel: x and v values (stacked subplots with shared x-axis)
ax2_top = fig.add_subplot(gs[0, 1])
ax2_bottom = fig.add_subplot(gs[1, 1], sharex=ax2_top)

# Top right: x from file2 (E=22.07, x_in=7.0)
ax2_top.plot(time_win2, x_vals2, color='#0059b3')
ax2_top.axhline(y=0.0, color="gray", linestyle="--")
ax2_top.axhline(y=3.693980625181293, color="red", alpha=0.5, linestyle="--")
ax2_top.axhline(y=7.7345908033901365, color="gray", linestyle="--")
ax2_top.text(0.03, 0.95, '(b)', transform=ax2_top.transAxes, fontsize=12, va='top')
ax2_top.set_ylabel(r'$\langle x \rangle$', fontsize=12)
ax2_top.margins(x=0.05, y=0.05)  # Add padding to ensure text visibility
ax2_top.set_ylim([-1, 11.0])

# Bottom right: v from file2 (E=22.07, x_in=7.0)
ax2_bottom.plot(time_win2, v_vals2, color='#0059b3')
ax2_bottom.text(0.03, 0.95, '(d)', transform=ax2_bottom.transAxes, fontsize=12, va='top')
ax2_bottom.set_ylabel(r'$\mathrm{Var}(x)$', fontsize=12)
ax2_bottom.set_xlabel(r'$t$', fontsize=12)
ax2_bottom.set_ylim([0.0, 5.8])
ax2_bottom.margins(x=0.05, y=0.05)  # Add padding to ensure text visibility

# Remove x-axis labels from top subplots
plt.setp(ax1_top.get_xticklabels(), visible=False)
plt.setp(ax2_top.get_xticklabels(), visible=False)
ax1_top.set_xlim([0, 100])
ax2_top.set_xlim([0, 100])

ax1_top.text(0.4, 1.2, r'\text{Left Well}', transform=ax1_top.transAxes, fontsize=12, va='top')
ax2_top.text(0.4, 1.2, r'\text{Right Well}', transform=ax2_top.transAxes, fontsize=12, va='top')

plt.tight_layout()

# Save the plot as a PDF
fig.savefig("fig3.pdf", bbox_inches="tight")
# Display the plot
plt.show()
```
